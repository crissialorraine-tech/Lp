```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("--- INICIANDO EXPERIMENTO DE ENGENHARIA DE DADOS ---")

datas = ['2021-12-31', '2022-12-31', '2023-12-31']
df = pd.DataFrame({
    'Data': pd.to_datetime(datas),
    'Media_Taxa': [9.5559, 9.9317, 11.2288],
    'Max_Taxa': [10.6792, 11.5636, 13.7085],
    'Min_Taxa': [8.6473, 8.9204, 8.3334],
    'Desvio_Padrao': [0.5376, 0.6854, 1.6180],
    'Var_Media_%': [0.0134, -0.0341, 0.1180]
})

print("\nTABELA 1: Resumo Estatístico Anual (Processamento Pandas/NumPy)")
print(df.to_string(index=False))

plt.figure(figsize=(10, 5))
datas_graf = pd.date_range('2021-01-01', '2023-12-31', freq='D')
np.random.seed(42)
taxa_observada = 10 + np.cumsum(np.random.randn(len(datas_graf)) * 0.05)
media_historica = np.full(len(datas_graf), 10.4)

plt.plot(datas_graf, taxa_observada, label='Taxa Observada', color='black')
plt.plot(datas_graf, media_historica, label='Média Histórica', color='red', linestyle='--')
plt.title('Evolução do Indicador Financeiro e Análise de Tendência (2021-2023)')
plt.ylabel('Valor do Indicador (%)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
